```python
import igraph 
from igraph import Graph, Plot
import pandas as pd

from rdkit import Chem
from rdkit.Chem import Draw

from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit.Chem import Descriptors

from collections import Counter
import pickle
```

```python
df_all_aglycons_with_single_taxonomy = pd.read_pickle("df_all_aglycons_with_single_taxonomy.pkl")
```

```python
#df_all_aglycons_with_single_taxonomy
```

---

```python
fpc095 = pd.read_pickle("fingerprint_comparison095.pkl")
```

```python
fpc095_graph = Graph.DataFrame(fpc095, directed=False)
```

```python
fpc095_cluster = Graph.components(fpc095_graph)
```

```python
fpc095_nodes = fpc095_graph.get_vertex_dataframe()
```

```python
multi_cluster_list095 = []
for index_list in fpc095_cluster:
    single_cluster_list = []
    for index in index_list:
        for ind, aglycon in enumerate(df_all_aglycons_with_single_taxonomy.deglycosilated_smiles):
            try:
                if aglycon == fpc095_nodes.name[index]:
                    aglycon_tax_pair = {aglycon: list(df_all_aglycons_with_single_taxonomy.taxonomy[ind])[0]}          
            except KeyError:
                print(index, ind, aglycon)
        single_cluster_list.append(aglycon_tax_pair)
    multi_cluster_list095.append(single_cluster_list)
```

```python
#multi_cluster_list095
```

```python
cnt = Counter()
for ind1, aglycon_tax_list in enumerate(multi_cluster_list095):
    cnt.clear()
    len_aglycon_tax_list = len(aglycon_tax_list)
    if len_aglycon_tax_list > 2:
        for aglycon_tax in aglycon_tax_list:
            for tax in aglycon_tax.values():
                if tax != 'no':
                    cnt[tax] += 1
        try:
            name_of_tax = max(cnt)
            quantity_of_tax = max(cnt.values())
            if quantity_of_tax >= len_aglycon_tax_list/2:
                for ind2, aglycon_tax in enumerate(aglycon_tax_list):
                    for aglycon in aglycon_tax:
                        for tax in aglycon_tax.values():
                            if tax == 'no':
                                 multi_cluster_list095[ind1][ind2][aglycon] = 'predict'+ '_' + name_of_tax
                            elif tax != name_of_tax:
                                 multi_cluster_list095[ind1][ind2][aglycon] = tax + '/' + name_of_tax
        except ValueError:
            pass
```

```python
#multi_cluster_list095
```

```python
multi_df_cluster_list095 = []
for aglycon_tax_list in multi_cluster_list095:
    df_cluster = pd.DataFrame(index=range(len(aglycon_tax_list)))
    df_cluster["aglycon"] = ""
    df_cluster["taxonomy"] = ""
    counter = 0
    for aglycon_tax in aglycon_tax_list:
        df_cluster.aglycon[counter] = list(aglycon_tax.keys())
        df_cluster.taxonomy[counter] = list(aglycon_tax.values())
        counter += 1
    multi_df_cluster_list095.append(df_cluster)
```

```python
#multi_df_cluster_list095
```

```python
cnt095 = Counter()
for dataframe in multi_df_cluster_list095:
    for dict_tax in dataframe.taxonomy:
        for tax in dict_tax:
            cnt095[tax] += 1
cnt095
```

```python
#for i in multi_df_cluster_list095:
#    display(i)
```

```python
with open("multi_df_cluster_list095.txt", "wb") as outfile:
    pickle.dump(multi_df_cluster_list095, outfile)
```
